```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("sales.csv")

print("Data:\n", data)

# Calculate average marks
average = data["Sales"].mean()
print("Average Sales per Year is :", average)

# Bar chart
plt.bar(data["Month"], data["Sales"])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales of a Product per Year")
plt.show()

```
